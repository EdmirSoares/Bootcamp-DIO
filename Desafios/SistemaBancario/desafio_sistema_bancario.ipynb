Desafio Sistema Bancário com Python


Operação de Depósito


```python
import time

saldo = {
    "saldo": 0,
    "extrato": [],
    "totalSaques": 0,
}

usuarios_cadastrados = []

contas_correntes = []


def insere_usuario():
    texto = " Cadastro de Usuário "
    print(texto.center(30, "-") + "\n")
    try:
        nome = input("Informe seu nome: ")
        data_nascimento = int(input("Informe sua data de nascimento: "))
        cpf = int(input("Informe os números do seu CPF: "))
        if cpf in [usuario["cpf"] for usuario in usuarios_cadastrados]:
            print("CPF já cadastrado!\n")
            return 0
        endereco = input(
            "Informe seu endereço no formato 'Logadouro - Bairro - Cidade/Sigla do Estado: "
        )
        conta_corrente_usuario = insere_conta_corrente()
        if conta_corrente_usuario == 0:
            return 0
        usuarios_cadastrados.append(
            {
                "nome": nome,
                "data_nascimento": data_nascimento,
                "cpf": cpf,
                "endereco": endereco,
                "conta_corrente": conta_corrente_usuario,
            }
        )
        print("Usuário cadastrado com sucesso!\n")
        return usuarios_cadastrados
    except Exception as e:
        print(f"Erro ao cadastrar usuário! {e}\n")
        return usuarios_cadastrados


def insere_conta_corrente():
    texto = " Cadastro de Conta Corrente "
    print(texto.center(30, "-") + "\n")
    try:
        numero_conta = int(input("Informe o número da conta corrente: "))
        agencia = "0001"
        for conta in contas_correntes:
            if conta["numero_conta"] == numero_conta:
                print("Conta corrente já cadastrada!\n")
                return 0
        else:
            contas_correntes.append(
                {
                    "numero_conta": numero_conta,
                    "agencia": agencia,
                }
            )
            print("Conta corrente cadastrada com sucesso!\n")
            return contas_correntes[-1]
    except Exception as e:
        print(f"Erro ao cadastrar conta corrente! {e}\n")
        return contas_correntes


def verificar_usuario():
    texto = " Login "
    print(texto.center(30, "-") + "\n")
    try:
        if not usuarios_cadastrados:
            print("Nenhum usuário cadastrado!\n")
            return False
        else:
            cpf = int(input("Informe o cpf do usuário: "))
            for indice, user in enumerate(usuarios_cadastrados):
                if user["cpf"] == cpf:
                    print(f"Usuário {cpf} logado com sucesso!\n")
                    return main(usuarios_cadastrados[indice])
                else:
                    print("Usuário não encontrado!\n")
    except Exception as e:
        print(f"Erro ao logar usuário! {e}\n")
        return False


def insere_saldo(saldo_inserido, inserir_extrato):
    texto = " Depósito "
    print(texto.center(30, "-") + "\n")
    try:
        insere_saldo = float(input("Informe o valor a ser inserido no saldo total: "))
        if insere_saldo > 0:
            saldo_inserido += insere_saldo
            inserir_extrato.append(f"Depósito: R$ {insere_saldo:.2f}")
            print(
                f"Saldo atual atualizado! O valor atual é de R$ {saldo_inserido:.2f}\n"
            )
            return saldo_inserido, inserir_extrato
        else:
            print("informe um valor válido!\n")
            return saldo_inserido, inserir_extrato
    except Exception as e:
        print(f"Insira um valor válido! {e}\n")
        return saldo_inserido, inserir_extrato


def realiza_saque(saldo_atual, quantidade_saques, inserir_extrato):
    try:
        texto = " Saque "
        print(texto.center(30, "-") + "\n")
        valor_saque = float(input("Informe o valor a ser sacado: "))

        if quantidade_saques == 3:
            print("Limite de saques diários atingido!\n")
            return saldo_atual, quantidade_saques, inserir_extrato

        if saldo_atual == 0:
            print("Saldo insuficiente!\n")
            return saldo_atual, quantidade_saques, inserir_extrato

        if (
            saldo_atual > 0
            and valor_saque < saldo_atual
            and valor_saque < 500
            and quantidade_saques < 3
        ):
            saldo_atual -= valor_saque
            quantidade_saques += 1
            inserir_extrato.append(f"Saque: R$ {valor_saque:.2f}")
            print(f"Saque realizado com sucesso no valor de R$ {valor_saque:.2f}\n")
            print(f"Saldo atualizado é de R$ {saldo_atual:.2f}\n")
            print(f"Total de saques realizados: {quantidade_saques}\n")
            return saldo_atual, quantidade_saques, inserir_extrato
        else:
            print("Valor inválido!")
            return saldo_atual, quantidade_saques, inserir_extrato

    except Exception as e:
        print(f"Insira um valor válido! {e}\n")
        return saldo_atual, quantidade_saques, inserir_extrato


def checar_extrato(saldo_atual, *, extrato_atual, total_saques):

    texto = " Extrato de Transações "
    print(texto.center(30, "-") + "\n")
    print(f"Seu saldo atual é de: R$ {saldo_atual:.2f}\n")
    print("Transações realizadas: \n")
    for item in extrato_atual:
        print(item)
    print("\n")
    print(f"Total de Saques realizados: {total_saques}\n")
    return saldo_atual, extrato_atual, total_saques


def main(usuario_logado):
    while True:
        print("Bem-vindo ao Banco Python!\n")
        print(f"Usuário logado: {usuario_logado['nome']}\n")
        print("Escolha uma opção: \n")
        print("2 - Inserir saldo")
        print("3 - Realizar saque")
        print("4 - Checar extrato")
        print("5 - Sair\n")
        opcao = int(input("Informe a opção desejada: "))

        if opcao == 2:
            saldo["saldo"], saldo["extrato"] = insere_saldo(
                saldo["saldo"], saldo["extrato"]
            )
        elif opcao == 3:
            saldo["saldo"], saldo["totalSaques"], saldo["extrato"] = realiza_saque(
                saldo["saldo"], saldo["totalSaques"], saldo["extrato"]
            )
        elif opcao == 4:
            saldo["saldo"], saldo["extrato"], saldo["totalSaques"] = checar_extrato(
                saldo["saldo"],
                extrato_atual=saldo["extrato"],
                total_saques=saldo["totalSaques"],
            )
        elif opcao == 5:
            print("Voltando ao Menu Principal...")
            time.sleep(2)
            break
        else:
            print("Opção inválida! Tente novamente.\n")


while True:
    opcao = input("1 - Cadastrar usuário\n2 - Login\n3 - Sair\n")
    if opcao == "1":
        insere_usuario()
    elif opcao == "2":
        verificar_usuario()
    elif opcao == "3":
        print("Saindo do sistema...")
        time.sleep(2)
        break
    else:
        print("Opção inválida! Tente novamente.\n")
```
